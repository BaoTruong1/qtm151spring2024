#Lecture 5

```python
num_simulations = 2000
sample_size_list = [1, 10, 50, 100, 200]
for sample_size in sample_size_list:
    vec_xbar=[None]*num_simulations
    for i in range(num_simulations):
        vec_a = np.random.chisquare(df=1, size=sample_size)
        vec_xbar[i]=vec_a.mean()
    plt.hist(vec_xbar)
    plt.title(str(sample_size))
    plt.show()
```

```python
data=pd.DataFrame()
data["age"]=[18,29,15,32,6]
data["num_underage_siblings"]=[0,0,1,1,0]
data["num_adult_siblings"]=[1,0,0,1,0]
display(data)
```

```python
fn_istwenties = lambda age: (age>20)&(age<30)
fn_sum=lambda x,y: x+y
def fn_agebracket(age):
    if(age>=18):
        return "Adult"
    elif(age<10):
        return "Child"
    else:
        return "Adolescent"
data["can_vote"]=list(map(fn_iseligblevote,data["age"]))
data["in_twenties"]=list(map(fn_istwenties,data["age"]))
data["age_brackets"]=list(map(fn_agebracket,data["age"]))
display(data)
```
